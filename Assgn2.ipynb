```python
#import library that need to use
import pandas as pd
import numpy as np
from matplotlib import pyplot as pt
```

```python
#read file
da = pd.read_csv('Malaysian_Tourism_Data_1995_to_2019.csv')
da
```

```python
#create total column to calculate sum
da['total'] = da.sum(axis = 1, numeric_only= True)
da
```

```python
newda = da.iloc[1:]
newda
```

```python
newda.to_csv('Tourism_Data_Malaysian_1995_to_2019_with_Total.csv')
```

```python
newda = pd.read_csv('Tourism_Data_Malaysian_1995_to_2019_with_Total.csv')
newda
```

```python
#declare data needed to plot the graph
#click view output fullscreen to see the details
x = list(newda['Country Code'])
y = list(newda['total'])
fig, ax = pt.subplots(figsize=(200,20))
pt.xlabel('Country Code')
pt.ylabel('total')
ax.plot(x,y, linewidth = 2)
pt.show()
```
